```python
import pandas as pd

df = pd.read_json("Potions.json")
df.index = df[0]
df = df[3]
potions = df.to_dict()
```

```python
potions_clean = dict()
for k,v in potions.items():
    if v is not None and v not in ['N.N', 'N.N.']:
        potions_clean[k] = v.split("\n")

potions_clean
```

```python
import json
with open("../Data/potions.json", "w") as outfile:
    json.dump(potions_clean, outfile)
```
